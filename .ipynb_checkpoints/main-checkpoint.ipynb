```python
# Read the presaved data
image_summaries = pd.read_csv('places_summaries.csv')
image_summaries['used_q'] = False
```

```python
# reference_image = mpimg.imread("kenzie-kraft-sWRNaS5Ijgg-unsplash.jpg")
# reference_image = cv2.imread("bulksplash-mrthetrain-b2tc2aWJKWg.jpg")
reference_image = cv2.imread("processed.jpeg")
reference_image = reference_image[33:-33, 22:-22, :]
reference_image_test = reference_image
plt.imshow(reference_image)
```

```python
res = 64

reference_image_x, reference_image_y = reference_image.shape[0] // res, reference_image.shape[1] // res 
reference_image_metrics = []

for x in tqdm(range(reference_image_x)):
    for y in range(reference_image_y):
        cropped_segment = reference_image[x*res:(x+1)*res, y*res:(y+1)*res, :]
        cropped_segment_averages = [np.mean(cropped_segment[:, :, i]) for i in range(3)]        
        reference_image_metrics.append([x*res, (x+1)*res, y*res, (y+1)*res] + cropped_segment_averages)
        
reference_image_metrics = pd.DataFrame(reference_image_metrics,
                                      columns=['x_start_index', 'x_end_index', 
                                              'y_start_index', 'y_end_index'] + color_columns)

reference_image_metrics['nearest_image'] = None
reference_image_metrics['nearest_image_summary'] = None
```

```python
test_n = 5000

for index, row in tqdm(reference_image_metrics.head(test_n).iterrows(), total=reference_image_metrics.head(test_n).shape[0]):
    neighbors = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(
        image_summaries[image_summaries.used_q == False][color_columns].values)
    distances, indices = neighbors.kneighbors([row[color_columns].values])
    nearest_index = indices[0, 0]
    reference_image_metrics.at[index, 'nearest_image'] = image_summaries.iloc[nearest_index].filename
    reference_image_metrics.at[index, 'nearest_image_summary'] = tuple(
        image_summaries.iloc[nearest_index][color_columns].values)
    image_summaries.at[nearest_index, 'used_q'] = True
    

# def find_nearest_images(index, row, image_summaries_in, reference_image_metrics_in):
#     neighbors = NearestNeighbors(n_neighbors=1, algorithm='ball_tree').fit(
#         image_summaries_in[image_summaries_in.used_q == False][color_columns].values)
#     distances, indices = neighbors.kneighbors([row[color_columns].values])
#     nearest_index = indices[0, 0]
#     reference_image_metrics_in.at[index, 'nearest_image'] = image_summaries_in.iloc[nearest_index].filename
#     reference_image_metrics_in.at[index, 'nearest_image_summary'] = tuple(
#         image_summaries_in.iloc[nearest_index][color_columns].values)
#     image_summaries_in.at[nearest_index, 'used_q'] = True

# Parallel(n_jobs=num_cores)(delayed(find_nearest_images)(index, row, image_summaries, reference_image_metrics) 
#                            for index, row in tqdm(reference_image_metrics.iterrows(), total=reference_image_metrics.shape[0]))
```
