```python
from glob import glob
from tqdm import tqdm
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

filenames = glob(r'C:\Users\Alcatraz\Downloads\archive\img_align_celeba\img_align_celeba\*')
```

```python
image_summaries = []

for filename in tqdm(filenames, total=len(filenames)):
    image = cv2.imread(filename)[37:-37, 17:-17, :]
    averages = [filename, np.mean(image[:, :, 0]), np.mean(image[:, :, 1]), np.mean(image[:, :, 2])]
    image_summaries.append(averages)
```

```python
image_summaries = pd.DataFrame(image_summaries, columns=['filename', 'r_average', 'g_average', 'b_average'])
```

```python
reference_image = mpimg.imread("kenzie-kraft-sWRNaS5Ijgg-unsplash.jpg")
reference_image = reference_image[33:-33, 22:-22, :]
plt.imshow(reference_image)
```

```python
reference_image_x, reference_image_y = reference_image.shape[0] // 72, reference_image.shape[1] // 72 
reference_image_metrics = []

for x in tqdm(range(reference_image_x)):
    for y in range(reference_image_y):
        cropped_segment = reference_image[x*72:(x+1)*72, y*72:(y+1)*72, :]
        cropped_segment_averages = [np.mean(cropped_segment[:, :, i]) for i in range(3)]        
        reference_image_metrics.append([x*72, (x+1)*72, y*72, (y+1)*72] + cropped_segment_averages)
        
reference_image_metrics = pd.DataFrame(reference_image_metrics,
                                      columns=['x_start_index', 'x_end_index', 
                                              'y_start_index', 'y_end_index', 
                                              'r_average', 'g_average', 'b_average']
                                      )
```
